```python
#LEGGERE I FILE
#******************************************

#in aggiunta al read si inserisce sep=';' per definire il separatore, in questo caso il punto e virgola
df_csv = pd.read_csv('Data/test.csv')
df_xls = pd.read_excel('Data/Superstore.xls')

#E' SEMPRE BUONA PRASSI DOPO IL LOADING DEL CSV DI CREARE UN BACKUP DEL DATAFRAME PER AVERE SEMPRE
#LA COPIA ORIGINALE DEI DATI
df_csv_backup = df_csv.copy()
df_xls_backup = df_xls.copy()
```

```python
#CANCELLARE LE COLONNE
#******************************************

#.drop cancellare colonne
#Prima faccio un array con le colonne da cancellare.
cols = ['Row ID', 'Order ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name', 'Quantity', 'Discount', 'Profit']

#Successivamente cancello le colonne dal dataframe, con df.drop(cols)
#Se axis è uguale a 0 si applica il metodo down sulle colonne, altrimenti se axis = 1 si applica il metodo across per ogni riga
#differenziare axis : righe e colonne, ha poco senso nella cancellazione, ha senso nella somma:
arr = np.array([[1,2,3],[4,5,6]])
print("ESEMPIO AXIS")
print (arr)
print("Axis 0: ")
print(np.sum(arr, axis=0))
print("Axis 1: ")
print(np.sum(arr, axis=1))
#Se inplace è falso allora assign your result to a new variable altrimenti inplace è true o need to assign

df_xls.drop(cols, axis=1, inplace=True)

df_xls.head()

#!!!! SE RIESEGUO MI DA ERRORE MA E' NORMALE IN QUANTO I DATI VENGONO CANCELLATI
```
